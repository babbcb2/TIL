```python
import re
import pandas as pd
```

```python
# 파일경로
def get_file(path):
    file = open(path, mode = 'r', encoding = "UTF-8")
    text = file.read()
    file.close()
    return text

# 정규표현식, \n제거
def get_re(text):
    regex_bracket = '\([^)]*\)'
    re_text = re.sub(regex_bracket, '', text)
    re_text = re_text.replace('\n','')
    return re_text


# 문장뒤에 들어가는데 붙어있는 경우 ex('."')
def get_replace_multi(re_text,*args):
    for i in args:
        multi_text = re_text.replace(i,"&&$")
    return multi_text

# 문장의 앞에 들어가는 기호
def get_replace_front(front_text,*frontargs):
    for i in frontargs:
        front_text = front_text.replace(i,"$"+i)
    return front_text

# 문장의 뒤에 들어가는 기호
def get_replace_back(back_text,*backargs):
    for i in backargs:
        print(i)
        back_text = back_text.replace(i,i+"$")
    return back_text




# 문장 분리 단위 전처리  (앞/뒤 구분이 불가능할때만 사용 위에 방식 권장)
def get_replace_unapart(re_text,*args):
    for i in args:
        unapart_text = re_text.replace(i,i+"$"+i)
    return unapart_text


# 문장분리($사용)
def get_split(unapart_text):
    text_list = unapart_text.split('$')
    return text_list

# 의미없는 행 제거 
def del_useless(text_list, del_list):
    for i in del_list:
        while True:
            try :
                text_list.remove(i)
            except :
                break
    return text_list     
```

```python
text = get_file('./picturebro.txt')
re_text = get_re(text)
# multi_text = get_replace_multi(re_text,".”")
front_text = get_replace_front(re_text,"“")
back_text = get_replace_back(front_text,"”",".")
# back_text = get_replace_back(front_text,"”")
# back_text = get_replace_back(front_text,".")

# 함수 아님
# back_text = back_text.replace('&&', '”')

text_list = get_split(back_text)
# unapart_text = get_replace_unapart(re_text,"“","”",".")
# text_list = get_split(unapart_text)


final_list = del_useless(text_list, ['',' ','  ' '.', "“","”"])
```
