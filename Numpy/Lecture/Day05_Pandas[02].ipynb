- reset_index () : 새로운 인덱스 할당, 기존 인덱스는 인덱스라는 새로운 피처로 추가할 수 있다.

```python
datasets_reset_idx_df = datasets_drop_df.reset_index (inplace=False)
```

```python
datasets_reset_idx_df
```

```python
datasets_reset_idx_df[['pclass', 'fare']].head()
```

```python
# pclass 3 인 데이터만 확인하고 싶다면?
datasets_reset_idx_df [datasets_reset_idx_df['pclass'] == 3]
```

```python
datasets_reset_idx_df [datasets_reset_idx_df['pclass'] == 3].iloc[0:10, [3,10]]
```

```python
# 나이가 60 이상인 정보만 추출하고 싶다면?
# pclass, survived, who

datasets_reset_idx_df [datasets_reset_idx_df['age_by_10'] > 60]

datasets_reset_idx_df.loc [datasets_reset_idx_df['age_by_10'] > 60, ['pclass', 'survived', 'who']]
```

- 여러개의 복합조건을 이용해서 불리언 인덱스를 만들어서 작업
- and -> &
- or -> |
- not -> ~

```python
# 나이가 60 보다 크고 선실등급 1등급이고 성별이 여자인 데이터 추출한다면?

age_over = datasets_reset_idx_df['age'] > 60
print (age_over)

pclass_equal = datasets_reset_idx_df['pclass'] == 1
print (pclass_equal)

gender = datasets_reset_idx_df['sex'] == 'female'
print (gender)

datasets_reset_idx_df [age_over & pclass_equal & gender]
```
