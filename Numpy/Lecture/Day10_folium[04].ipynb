```python
seoul_school_df = pd.read_excel('./data/elem_school_folium.xlsx')
seoul_school_df.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명주소, 위도, 경도

seoul_school_df = seoul_school_df.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
seoul_school_df.head()
```

```python
# 소재지 도려명주소 서울을 포함하고 학교구분이 초등학교만 추출

seoul_school_df.isnull().sum()
```

```python
seoul_school_df.dropna(inplace=True)
seoul_school_df.isnull().sum()
```

```python
# 소재지 도려명주소 서울을 포함하고 학교구분이 초등학교만 추출
map_df = seoul_school_df[seoul_school_df['소재지도로명주소'].str.contains('서울특별시') &
                    seoul_school_df['학교급구분'].str.contains('초등학교')]
map_df
```

```python
seou_ele_map = f.Map(location=[37.56666466213293, 126.97842615527017], zoom_start=12)

for idx in map_df.index :
    f.CircleMarker([ map_df.loc[idx, '위도'],  map_df.loc[idx, '경도']],
            popup = map_df.loc[idx, '학교명'],
            color = 'brown', # 원의 둘레 색상
            fill = True,    # 원안의 색상
            fill_color = 'coral').add_to(seou_ele_map)
    
    
#    f.Marker([ map_df.loc[idx, '위도'],  map_df.loc[idx, '경도']],
#            popup = map_df.loc[idx, '학교명'],
#            color = 'red',
#            fill = True,
#            fill_color = 'coral').add_to(seou_ele_map)

seou_ele_map
```
