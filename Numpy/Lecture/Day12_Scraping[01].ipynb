```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import HTTPError
from urllib.request import URLError
```

```python
try :
    html = urlopen('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx')
except HTTPError as he :
    print('Http error')
except URLError as ue :
    print('Url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser')
```

```python
table = soup.find('table', {'class' : 'tData01 tt'})
table
```

```python
# 선수명, AVG, H, HR, RBI
# player, bat_avg, hit, homerun, rbi
# csv 파일 저장# csv 파일 저장
# 시각화

# DB 입력

# 위정보를 가져와서 LIST 에 담고 싶다위정보를 가져와서 LIST 에 담고 싶다
```

```python
table.find ('tr')
```

```python
table.find_all ('tr')
```

```python
table.find_all ('td')
```

```python
# 선수명, AVG, H, HR, RBI
data = []

for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    for td in tds :
        if td.find('a') :
            #player_list.append(tds[1].text)
            data.append([tds[1].text ,tds[4].text, tds[8].text, tds[11].text, tds[13].text])
```

```python
data
```

```python
with open ('KBO_list.csv', 'w', encoding='utf-8') as file :
    file.write ('player, bat_average, hit, homerun, RBI\n')   # 컬럼에 이름 추가
    for d in data :
        file.write('{},{},{},{},{}\n'.format(d[0],d[1],d[2],d[3],d[4]))  # 데이터를 반복하여 줄 단위로 저장
print('file save !! ')
```

```python
# 영화명, 평점을 담는 리스트를 생성하라
# list comprehension

len(soup.find_all('td','point'))
```

```python
movie_list = [soup.find_all('div', 'tit5')[n].a.string for n in range (0, 50)]
point_list = [soup.find_all('td', 'point')[n].string for n in range (0, 50)]
```

```python
point_list
```

```python
# 날짜 생성 (50) 오늘날짜를 기준으로 50일
# pd.date_range()
# 날짜 , 영화명, 포인트



date_list = pd.date_range('2021-3-10', periods = 50)

movieDF = pd.DataFrame ({
    '날짜' : date_list,
    '영화명' : movie_list,
    '포인트' : point_list
})

movieDF.info()
```

```python
movieDF['포인트'] = movieDF['포인트'].astype(float)
movieDF.info()
```

```python
movieDF.sort_values(by='포인트', ascending=False, inplace=True)
```
