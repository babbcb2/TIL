```python
# file read option
# - path
# - setp
# - header
# - skiprows
# - encoding
# - fillna

cityDF = pd.read_excel ('./data/시도별 전출입 인구수.xlsx', header = 0)
```

```python
cityDF.head()
```

```python
cityDF = cityDF.fillna(method='ffill')
```

### df.fillna('바꾸길 원하는 값',inplace = True)

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋을 만든다면?
mask = (cityDF['전출지별'] == '서울특별시') & (cityDF['전입지별'] != '서울특별시')
seoulDF = cityDF[mask]                          
```

```python
seoulDF
```

```python
seoulDF.drop(['전출지별'], axis = 1, inplace=True)
seoulDF
```

```python
# 원하는 열 인덱스를 다름이름으로 변경
seoulDF.rename({'전입지별': '전입지'}, axis = 1, inplace=True)
seoulDF
```

```python
seoulDF.set_index('전입지', inplace = True)
seoulDF
```

```python
# 경기도 이동한 인구 데이터값만 선택해서 시각화?
datas = seoulDF.loc['경기도']
print(datas.index)
print(datas.values)
```

```python
plt.figure(figsize=(15,5))

plt.plot(datas.index, datas.values, marker='o', markersize=5)

plt.title ('서울 -> 경기 인구 이동')
plt.xlabel ('년도', size =20)
plt.ylabel ('인구 이동수', size=15)
plt.xticks(rotation='vertical')

plt.legend(labels=['서울->경기'], loc='best')

plt.show()
plt.close()
```

```python
plt.plot(datas)
```

```python
fig = plt.figure(figsize=(10,10))
ax01 = fig.add_subplot(2,1,1)
ax02 = fig.add_subplot(2,1,2)


#marker = o, +, *, .
ax01.plot(datas, marker='*', markersize=10)
ax02.plot(datas, marker='o', markersize=5, color='red', linewidth=2, label='서울->경기', markerfacecolor='black')
ax02.legend(loc='best')


# y축 범위를 지정
ax01.set_ylim(50000, 700000)
ax02.set_ylim(50000, 800000)

# label 각도
ax02.set_xticklabels(datas.index, rotation=75)

plt.show()
plt.close()
```

```python
# 서울에서 충청남도, 경상북도, 강원도 이동한 인구 데이터값만 선택
# 데이터프레임으로 만들어보자
# map()

col_years = list(map(str, range(1970, 2018)))

dataDF = seoulDF.loc[['충청남도', '경상북도', '강원도'], col_years]
dataDF
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years, dataDF.loc['충청남도', :], marker='o', label='서울->충남')

ax.plot(col_years, dataDF.loc['경상북도', :], marker='*', label='서울->충북')

ax.plot(col_years, dataDF.loc['강원도', :], marker='+', label='서울->경기')

ax.set_title('타이틀')

ax.set_xlabel('년도')
ax.set_ylabel('인구수')

ax.legend(loc='best')

plt.show()
plt.close()
```

```python
import seaborn as sns
iris_datasets = sns.load_dataset('iris')
iris_datasets.head()
```

```python
iris_datasets.index
```

```python
iris_datasets.columns
```

```python
iris_datasets['species'].value_counts()
```

```python
iris_datasets.rename(columns = {iris_datasets.columns[0]: 'SL',
                     iris_datasets.columns[1]: 'SW',
                     iris_datasets.columns[2]: 'PL',
                     iris_datasets.columns[3]: 'PW',
                     iris_datasets.columns[4]: 'Y'}, inplace=True)
```

```python
iris_datasets
```

```python
iris_datasets[['SL', 'SW']][ : 5]
```

```python
# 종을 기준으로 그룹화하여 각 그룹의 평균을 구해 본다면?
iris_mean_df = iris_datasets.groupby('Y').mean()
```

```python
iris_mean_df.T
```

```python
plt.figure()

iris_mean_df.plot(kind = 'bar',
                   figsize = (15,10))

plt.title('평균')
plt.xlabel('변수')
plt.ylabel('평균')

plt.show()
plt.close()
           
```

```python
plt.figure()

iris_mean_df.T.plot(kind = 'bar',
                   figsize = (15,10))

plt.title('평균')
plt.xlabel('변수')
plt.ylabel('평균')

plt.show()
plt.close()
           
```
