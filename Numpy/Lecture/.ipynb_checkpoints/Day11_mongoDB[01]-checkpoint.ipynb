```python
# pip install pymongo

import pymongo as mongo
```

```python
articleCollection = articleDB['article']
print (articleCollection)
```

## 3. Document insert
- insert_one() : inserOne()(mongo shell)
- insert_many() : insertMany()(mongo shell)

```python
doc01 = {'author' : 'jslim',
         'subject' : '드디어 몽고',
         'content' : '잉~~ 쉽네',
         'tags' : ['mongDB', 'python', 'pymong'],
         'view' : 50}
```

```python
articleCollection.insert_one(doc01)
```

## 4. Documnet find

- fine()

```python
doc03 = {'author' : 'changho',
         'subject' : '인공지능',
         'content' : '비정형 데이터 학습',
         'tags' : ['mongDB', 'pymong'],
         'view' : 200}
```

```python
articleCollection.insert_one(doc03)
```

```python
cursor = articleCollection.find( {'author' : 'changho' , 'view': {'$gte' : 200 } } )
for doc in cursor:
    print(doc)
```

```python
cursor = articleCollection.find( {'author' : 'changho' , 'view': {'$gte' : 200 } }, {'author' : 1}  )
for doc in cursor:
    print(doc)
```

```python
# update
articleCollection.update_one({'author' : 'jslim'}, {'$set' : {'view' : 150}})
```

```python
cursor = articleCollection.find()
for doc in cursor:
    print(doc)
```

```python
# remove
# delete_one ()
# delete_many ()

articleCollection.delete_many( {'author' : 'jslim'})
```

## Web Scraping- BeautifulSoup, Crawling - Selenium
- Web Scraping  : HTML, CSS와 같은 웹 페이지에서 구문분석(Parsing)을 거쳐 필요한 정보를 추출하는 방법
- BeautifupSoup : HTML 태그를 Parsing 해서 필요한 데이터만 추출하는 함수를 제공하는 python 라이브러리

```python
import requests

webpage = requests.get('https://www.daangn.com/hot_articles')
print(webpage.text)
```

```python
# BeautifulSoup 객체 생성 - html.parser 구문분석기

from bs4 import BeautifulSoup

webpage = requests.get('https://www.daangn.com/hot_articles')
soup    = BeautifulSoup(webpage.content , 'html.parser')
print (soup)
```

```python
print(soup.p) # 첫번째 만나는 p 태그를 가져옴
```

```python
print(soup.p.string) # 첫번째 만나는 p 태그의 String을 가져옴
```

```python
print(soup.h1) # 첫번째 만나는 h1 태그를 가져옴
```

```python
for a_tags in soup.h1.children :
    for child in a_tags :
        print (child)
```

```python
print (soup.ul) # 첫번째 만나는 ul 태그를 가져옴
```

```python
for parent in soup.ul.parents :
    print (parent)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import HTTPError
from urllib.request import URLError
```

```python
try :
    html = urlopen('https://pythondojang.bitbucket.io/weather/observation/currentweather.html')
except HTTPError as he :
    print('Http error')
except URLError as ue :
    print('Url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser')
```

```python
table = soup.find('table', {'class' : 'table_develop3'})
table
```

```python
for tr in table.find_all('tr') :
    #print(tr.find_all('td'))
    tds = tr.find_all ('td')
    for td in tds:
        print(td)
    print()
```

```python
for tr in table.find_all('tr') :
    #print(tr.find_all('td'))
    tds = tr.find_all ('td')
    for td in tds:
        if td.find('a') :
              print(td.find('a').text)
    print()
```

```python

temp_list = []
for tr in table.find_all('tr') :
    #print(tr.find_all('td'))
    tds = tr.find_all ('td')
    for td in tds:
        #if td.find('a') :
        #      print(td.find('a').text)
        # print(tds[5].text)
        temp_list.append (tds[5].text)
```

```python
temp_list
```

```python
point_list    = []
temp_list     = []
humidity_list = []
data = []
for tr in table.find_all('tr'):
    # print( tr.find_all('td') )
    tds = tr.find_all('td')
    for td in tds : 
        if td.find('a') :
            # print( td.find('a').text)
            # point_list.append( td.find('a').text )
            # print(tds[5].text)
            # temp_list.append( tds[5].text )
            # print(tds[9].text)
            # humidity_list.append(tds[9].text)
            data.append([td.find('a').text  , tds[5].text , tds[9].text])
```

```python
# point_list
# temp_list
# humidity_list
data
```

```python
# csv 파일 저장
# 2차원의 매트릭스 -> 1차원으로 가공하여 빼낸다

with open('python_weather.csv', 'w', encoding='utf-8') as file :
    for d in data :
        file.write ('{},{},{}\n'.format(d[0],d[1],d[2]))
print('file save ok~~~')
```
