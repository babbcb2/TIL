```python
from glob import glob
```

```python
files = glob('./data/지역*.xls')
tmp_row = []
for file in files :
    tmp = pd.read_excel(file, header = 2)
    tmp_row.append (tmp)
tmp_row
```

```python
oil_station = pd.concat(tmp_row)
oil_station
```

```python
oil_station.info()
```

```python
# 휘발유를 대상으루 분석
#         상호, 주소, 휘발유, 셀프여부, 상표


station = oil_station.filter(['상호','주소', '휘발유','셀프여부', '상표'])
station
```

```python
# rename  oil_store, address, price, self_y_n, company

station.rename(columns = {'상호' : 'oil_store',
                                 '주소' : 'address',
                                 '휘발유' : 'price',
                                 '셀프여부' : 'self_y_n',
                                 '상표' : 'company'}, inplace = True)

station
```

```python
# 파생변수 생성 (direct)
# 구별 주유소 가격을 조사하기 위해서

station['direct'] = [addr.split()[1] for addr in station['address']]
station.head()
```

```python
# 구별정보확인 (unique())
# 서울특별시 -> 성동구
# 특별시 -> 도봉구

# 밸류값 확인 = station['direct'].unique()

station.loc[station['direct'] == '서울특별시', 'direct']  = '성동구'
station.loc[station['direct'] == '특별시', 'direct']  = '도봉구'
station['direct'].unique()
```

```python
# 구별정보확인 (unique())
```

```python
# 가격 정보 확인[ - 들어있음] 후
# 가격 정보가 기입되지 않은 주유소는 제거
# 컬럼 타입을 숫자형으로 변경

station = station[station['price'] != '-']
station
station['price'] = [int(price) for price in station['price']]
station.info
```

```python
# 인덱스를 새롭게 정의(reset_index)
# 인덱스열을 제거

station.reset_index (inplace = True)
station
```

```python
# 박스플롯을 이용한 셀프여부 비교 시각화

fig = plt.figure(figsize = (7, 5))

#sns.boxplot(x='self_y_n', y='price', data = station, palette = 'Set3')
sns.swarmplot(x='self_y_n', y='price', data = station, palette = 'Set3')

plt.show()
plt.close()
```

```python
# 회사별 가격 시각화

fig = plt.figure(figsize = (7, 5))
sns.boxplot (x='company', y='price', hue='self_y_n', data=station)
plt.show()
plt.close()
```

```python
# 구별 휘발유 평균가격 확인

direct_pivot = pd.pivot_table(station , index='direct' , values='price' , aggfunc=np.mean)
direct_pivot.head()
```

```python
geo_str = json.load(open('./data/02. skorea_municipalities_geo_simple.json', encoding='utf-8'))
geo_str
```

```python
map = f.Map(location=[37.55, 126.98] , zoom_start=10.5)
f.Choropleth(geo_data = geo_str , 
             data = direct_pivot ,
             columns = [direct_pivot.index , 'price'] , 
             key_on='feature.id').add_to(map)
map
```

```python
# 지도 시각화를 위해서
# 주유가격 상위 10개, 하위 10개
station_tap10    = station.sort_values(by='price', ascending=False).head(10)
station_bottom10 = station.sort_values(by='price', ascending=True).head(10)

display (station_tap10)
display (station_bottom10)
```

```python
gmap_key = 'AIzaSyByteMdsP-K7ztjTp5e-iqOqhBT4Ax__Sk'
gmaps = googlemaps.Client(key = gmap_key)
```

```python
top_lat = []
top_lng = []

for idx in station_tap10.index :
    #print (idx)
    
    tmp_addr = station_tap10['address'][idx]
    #print(tmp_addr)
    
    tmp_map = gmaps.geocode(tmp_addr)
    #print(tmp_map)
    
    tmp_loc = tmp_map[0].get('geometry')
    top_lat.append(tmp_loc['location']['lat'])
    top_lng.append(tmp_loc['location']['lng'])
    
print(top_lat)
print(top_lng)
```

```python
bottom_lat = []
bottom_lng = []

for idx in station_bottom10.index :
    #print (idx)
    
    tmp_addr = station_bottom10['address'][idx]
    #print(tmp_addr)
    
    tmp_map = gmaps.geocode(tmp_addr)
    #print(tmp_map)
    
    tmp_loc = tmp_map[0].get('geometry')
    bottom_lat.append(tmp_loc['location']['lat'])
    bottom_lng.append(tmp_loc['location']['lng'])
    
print(bottom_lat)
print(bottom_lng)
```

```python
station_tap10['lat'] = top_lat
station_tap10['lng'] = top_lng

station_bottom10['lat'] = bottom_lat
station_bottom10['lng'] = bottom_lng

display (station_tap10)
display (station_bottom10)
```

```python
for idx in station_tap10.index :
    f.Marker([station_tap10.loc[idx,'lat'], station_tap10.loc[idx,'lng'] ],
            popup = station_tap10.loc[idx, 'oil_store'],
            #icon = f.icon(color='red')
            ).add_to(map)

for idx in station_bottom10.index :
    f.Marker([station_bottom10.loc[idx,'lat'], station_bottom10.loc[idx,'lng'] ],
            popup = station_bottom10.loc[idx, 'oil_store'],
            #icon = f.icon(color='blue')
            ).add_to(map)    
map
```
